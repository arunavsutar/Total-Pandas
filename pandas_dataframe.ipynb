```python
#for dataframe we aplly similarly
points_df = pd.DataFrame(
    {
        '1st point':[(3,4),(-6,5),(0,0),(-10,1),(4,5)],
        '2nd point':[(-3,4),(0,0),(2,2),(10,10),(1,1)]
    }
)

points_df
```

```python
def euclidean(row):
  pt_A = row['1st point']
  pt_B = row['2nd point']

  return ((pt_A[0] - pt_B[0])**2 + (pt_A[1] - pt_B[1])**2)**0.5
```

```python
points_df['distance'] = points_df.apply(euclidean,axis=1)
points_df
```

## Merge , Concat Operations

```python
courses = pd.read_csv('courses.csv')
students = pd.read_csv('students.csv')
nov = pd.read_csv('reg-month1.csv')
dec = pd.read_csv('reg-month2.csv')
matches = pd.read_csv('matches.csv')
delivery = pd.read_csv('deliveries.csv')
```

```python
# pd.concat
# df.concat
# ignore_index
# df.append
# mullitindex -> fetch using iloc
# concat dataframes horizontally
```

```python
#When merging the index got duplicated so to remove that we use
#ignore_index=True
regs = pd.concat([nov,dec],ignore_index=True)
```

```python
regs
```

```python
nov.append(dec,ignore_index=True)
#These both syntax are similar
```

```python
#MultiIndex DataFrame
multi = pd.concat([nov,dec],keys=['Nov','Dec'])
```

```python
pd.concat([nov,dec],axis=1)
#Side by side concat i.e row wise concat or Horizontally concat
```

```python
#Merging
```

```python
#inner join
students.merge(regs,how='inner',on='student_id')
#for different column name in different table write
#left_on= and right_on=
#how is used to write the type of join or merge
#by default how='inner'
```

```python
# left join
courses.merge(regs,how='left',on='course_id')
```

```python
temp_df=pd.DataFrame({
    'student_id':[26,27,28],
    'name':['Nitish','Ankit','Rahul'],
    'partner':[28,26,17]
})
students=pd.concat([students,temp_df],ignore_index=True)
```

```python
students.tail()
```

```python
#right join
students.merge(regs,how='right',on='student_id')
```

```python
# outer join
students.merge(regs,how='outer',on='student_id').tail(10)
```

```python
#  Plot bar chart for revenue/course
regs.merge(courses,on='course_id').groupby('course_name')['price'].sum().plot(kind='bar')
```

```python
#  Print student name -> partner name for all enrolled students
# self join
students.merge(students,how='inner',left_on='partner',right_on='student_id')[['name_x','name_y']]
```

```python
# Alternate syntax for merge
# students.merge(regs)

pd.merge(students,regs,how='inner',on='student_id')
```
